```python
# Function to get nearest Monday

def get_nearest_monday(date):
    if date.weekday() == 5:  # Saturday
        return date + pd.Timedelta(days=2)  # Add 2 days to get Monday
    elif date.weekday() == 6:  # Sunday
        return date + pd.Timedelta(days=1)  # Add 1 day to get Monday
    else:
        return date  # Weekday, no change
```

```python
date = pd.to_datetime('2019-01-16')

# Get the weekday (0 = Monday, 6 = Sunday)
weekday = date.weekday()

# Print the weekday
print(weekday)
```
